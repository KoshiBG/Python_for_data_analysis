# Завдання 1
**Скільки різних паролів можна скласти, використовуючи три літери англійського алфавіту, якщо літери не повинні повторюватися?**

Англійський алфавіт містить 26 літер.
Для створення пароля з трьох літер, причому без повторень, ми можемо використовувати перестановки, оскільки важливий порядок літер.
Першу літеру можна вибрати з 26 можливих літер.
Другу літеру можна вибрати з 25 літер (оскільки одна вже вибрана).
Третю літеру можна вибрати з 24 літер (дві вже вибрані).
Формула:
Кількість перестановок трьох літер з 26 дорівнює:
P(26,3)=26×25×24

Обчислення:
26×25×24=15600

Кількість різних паролів, які можна скласти з трьох різних літер англійського алфавіту, дорівнює 15600.

# Завдання 4
В урні знаходяться 4 червоні, 3 сині і 2 зелені кулі. Скількома способами можна витягти 3 кулі, якщо вони повинні бути різних кольорів?

Щоб знайти кількість способів витягти 3 кулі з урни, при цьому всі кулі повинні бути різних кольорів (червона, синя і зелена), ми використовуємо правило добутку. Оскільки кількість куль кожного кольору є обмеженою, і для кожного кольору ми вибираємо лише одну кулю, то рішення виглядає так:

Кількість червоних куль: є 4 червоні кулі, тому можемо вибрати 1 червону кулю 4 способами.

Кількість синіх куль: є 3 сині кулі, тому можемо вибрати 1 синю кулю 3 способами.

Кількість зелених куль: є 2 зелені кулі, тому можемо вибрати 1 зелену кулю 2 способами.

Загальна кількість способів:
За правилом добутку кількість способів вибрати 3 кулі різних кольорів обчислюється як:

Кількість способів=4×3×2=24

Кількість способів витягти 3 кулі, при цьому всі кулі різних кольорів, дорівнює 24.
