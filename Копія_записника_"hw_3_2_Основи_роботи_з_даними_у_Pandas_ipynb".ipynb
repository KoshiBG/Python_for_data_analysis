```python
# Фільтрація даних для міст Yangon та Mandalay
filtered_data = df[df['City'].isin(['Yangon', 'Mandalay'])]

# Обчислення середнього значення gross income для кожного міста
average_gross_income = filtered_data.groupby('City')['gross income'].mean()

# Виведення результатів
print(average_gross_income)
```

```python
average_gross_income/average_gross_income.sum()
```

В місті Mandalay дохід більше на 1%. На цей чинник можуть впливати наступні наявні данні:
1.Branch (Філія): Різні філії можуть мати різні обсяги продажів, залежно від їхнього розташування, розміру та кількості клієнтів.
2.City (Місто): Розташування магазину в різних містах може впливати на валовий дохід через відмінності в попиті, купівельній спроможності клієнтів і конкуренції.
3.Customer type (Тип клієнта): Різні типи клієнтів (наприклад, нові або постійні клієнти) можуть мати різні купівельні звички, що впливає на загальний дохід
4.Gender (Стать): Можливо, стать клієнтів може впливати на типи продуктів, які вони купують, і їх загальну вартість покупок.
5.Product line (Продуктова лінійка): Різні продуктові лінійки можуть мати різні рівні популярності та цінові категорії, що безпосередньо впливає на валовий дохід.
6.Unit price (Ціна за одиницю): Ціна продукту безпосередньо впливає на валовий дохід, оскільки більші ціни зазвичай означають більший дохід за одиницю товару.
7.Date (Дата): Продажі можуть варіюватися в залежності від днів тижня, місяців або сезонів, що впливає на валовий дохід.
8.Time (Час): Час доби може впливати на кількість продажів, наприклад, певні години можуть бути більш популярними для покупок.
9.Payment (Спосіб оплати): Різні способи оплати можуть впливати на зручність покупок і, як наслідок, на валовий дохід.

9. Врахувуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.

```python
# Відфільтруємо дані для міст Yangon та Mandalay

filtered_data = df[df['City'].isin(['Yangon', 'Mandalay'])]

# Порахуємо кількість інвойсів по кожній категорії товарів для кожного міста

invoice_counts = filtered_data.groupby(['City', 'Product line']).size()

#Обчислення відсоткового розподілу

percentage_distribution = invoice_counts/invoice_counts.sum()*100
print(percentage_distribution)

```
